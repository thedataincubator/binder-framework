```python
# Load Dataset
dataset = pd.read_csv('/content/drive/MyDrive/Data_incubator/data_sets/NYC_energy.csv')
display(dataset.shape)
display(dataset.head(10))
display(dataset.tail(10))
dataset.columns
```

```python
#  Lets write one function to find out 
## missing values ,unique values ,data types  --> EDA
def df_summary(df):
    df_U = df.nunique()
    df_M = df.isnull().sum()
    df_I = df.dtypes
    
    ## converting all data to dataframe
    df_U = df_U.to_frame().reset_index()
    df_M = df_M.to_frame().reset_index()
    df_I = df_I.to_frame().reset_index()
    
    ## renaming columns from default 0 to some sensible name
    df_U = df_U.rename(columns= {0: 'Unique Data'})
    df_M = df_M.rename(columns= {0: 'Missing Data'})
    df_I = df_I.rename(columns= {0: 'Data Types'})
    
    ## concatting the 3 dataframes. Remember pd.merge can merge only 2 df at a time
    output = pd.merge(pd.merge(df_M,df_U,on='index'),df_I,on='index')
    
    return output;  
```

```python
df_summary(dataset)
```

## Considering only NTAs with at least 5 listed businesses, what is the average total savings and the total jobs created for each NTA?

```python
# Filter the data for NTAs with at least 5 businesses
grouped_by_NTA = dataset.groupby('NTA').filter(lambda x: len(x) >= 5)

# Group the filtered data by NTA and calculate the average savings and total jobs created
grouped_by_NTA = grouped_by_NTA.groupby('NTA').agg({'Total Savings': 'mean', 'Job created': 'sum'})

# Print the result
print(grouped_by_NTA)
```

### 1. Create a scatter plot of jobs created versus average savings. Use both a standard and a logarithmic scale for the average savings.



```python
# Load dataset grouped by NTA that was saved on day 4
dataset2 = pd.read_csv('/content/drive/MyDrive/Data_incubator/data_sets/grouped_by_NTA.csv')
display(dataset2.shape)
display(dataset2.head(10))
display(dataset2.tail(10))
dataset2.columns
```

### Standard

```python
from sklearn.preprocessing import StandardScaler
# Select the columns to standardize
cols = ['Total Savings', 'Job created']

# Create a StandardScaler object
scaler = StandardScaler()

# Fit the scaler to the selected columns
scaler.fit(dataset2[cols])

# Transform the selected columns using the scaler
cols_standardized = scaler.transform(dataset2[cols])

# Convert the standardized columns to a DataFrame
df_standardized = pd.DataFrame(cols_standardized, columns=[f'{col}_standardized' for col in cols])

# Concatenate the original DataFrame and the standardized columns
df_concatenated = pd.concat([dataset2, df_standardized], axis=1)
display(df_concatenated.shape)
display(df_concatenated.head(10))
display(df_concatenated.tail(10))
df_concatenated.columns
```

```python
# Create a scatter plot using seaborn
plt.figure(figsize=(20, 10))
ax = sns.scatterplot(data=df_concatenated, x='Total Savings_standardized', y='Job created_standardized')

# Set the title and axis labels
ax.set(title='Standard Scatter Plot of Jobs created vs average total savings in NYC', xlabel='Number of Jobs created', ylabel='Standard Average total savings')
```

```python
# creating log scaled DataFrame
cols = ['Total Savings', 'Job created']
# Apply the log10 transformation to the selected columns
df_log = df_concatenated[cols].apply(np.log10)

# Rename the columns in the transformed DataFrame
df_log.columns = [f'{col}_log' for col in cols]

# Add the transformed columns to the original DataFrame
LogScaled_df = df_concatenated.assign(**df_log)
display(LogScaled_df.shape)
display(LogScaled_df.head(10))
display(LogScaled_df.tail(10))
LogScaled_df.columns
```

```python
# Scatter plot  vs  after log10 scaling
# Create a scatter plot using seaborn
plt.figure(figsize=(20, 10))
ax = sns.scatterplot(data=LogScaled_df, x='Job created_log', y='Total Savings_log')

# Set the title and axis labels
ax.set(title='Scatter Plot of Jobs created vs Log10 average total savings in NYC', xlabel='Number of Jobs created', ylabel='Average Log10 Scaled total savings')
```
