```python
string='yarrak@gmail.com'
string.split('@')[-1].strip('.com')
```

```python
"""
We define a function that strips the domain name from the email address.

We had to typecast into str since for some reason it was formatted
as float.
"""

def domain(row):
    string=str(row['company email'])
    string=string.split('@')[-1].strip('.com')
    row['company email']=string
    return row

df.apply(domain,axis=1)['company email'].value_counts()

#len(df.apply(domain,axis=1)['company email'].value_counts())
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
"""
I used total savings here instead of average,
how to define average?

plot.scatter(self, x, y, s=None, c=None, **kwargs)
s-> size
c-> colors
"""

x=df['Job created']
y=df['Total Savings']

plt.title('Job created vs Total Savings')
plt.xlabel('Job created')
plt.ylabel('Total Savings')

plt.scatter(x, y)
plt.show()
```

```python
fig = plt.figure()
ax = plt.gca()
ax.scatter(x ,y , edgecolors='none')
plt.title('Job created vs Total Savings ')
plt.xlabel('Job created')
plt.ylabel('Total Savings (log)')
ax.set_yscale('log')
```
